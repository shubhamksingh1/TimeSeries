# Random Walk

## Imports

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

## Set Fig Size

```python
plt.rcParams['figure.figsize'] = [10, 7.5]
```

## Generate Random Walk Data

```python
steps = np.random.standard_normal(1000)
steps[0]=0
random_walk = np.cumsum(steps)
```

```python
random_walk[:10]
```

## Plot Random Walk

```python
plt.plot(random_walk)
plt.title("Simulated Random Walk")
plt.show()
```

There is clear trend in the plot

## Plot Random Walk Coeff

```python
random_walk_acf_coef = acf(random_walk)
random_walk_acf_coef
```

## Plot ACF

```python
plot_acf(random_walk, lags=20);
```

As Clear from the plot there is trend in Random Walk, also emplies that data is not stationary

Notes:
 A time series with cyclic behaviour (but with no trend or seasonality) is stationary. This is because the cycles are not of a fixed length, so before we observe the series we cannot be sure where the peaks and troughs of the cycles will be.

In general, a stationary time series will have no predictable patterns in the long-term. Time plots will show the series to be roughly horizontal (although some cyclic behaviour is possible), with constant variance.

## Extracting Noise from Random Walk

```python
random_walk_diff = np.diff(random_walk, n=1)
```

```python
plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(random_walk_diff)
plt.title('Noise')
plt.show()
```

## Plotting ACF for Noise

```python
plot_acf(random_walk_diff, lags=20);
```

As Expected both the plots of noise emplies there is no trends in noise and Noise is stationary
